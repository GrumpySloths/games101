```python
from PIL import Image, ImageFilter

# 
image = Image.open("./example.png").convert("RGBA")

# 
shadow = Image.new(mode="RGBA", size=image.size)

# 
shadow_offset = (40, 40)
blur_radius = 10

# 
shadow.paste(image, shadow_offset)

# 
shadow = shadow.filter(ImageFilter.GaussianBlur(blur_radius))
print(shadow.mode)
# 
result = Image.blend(shadow, image,0.9)
result
# shadow

# 
# result.save("example_shadow.jpg")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
fs = 1000
N = 1000

# 
t = np.arange(N) / fs
x = np.sin(2 * np.pi *  t) + np.sin(3 * np.pi * t)

# 
fc = 80
h = 2 * fc / fs * np.sinc(2 * fc * t)
y = np.convolve(x, h)

# 
plt.plot(t, x, label='Original Signal')
plt.plot(t, y[:N], label='Filtered Signal')
plt.legend()
plt.show()
```
